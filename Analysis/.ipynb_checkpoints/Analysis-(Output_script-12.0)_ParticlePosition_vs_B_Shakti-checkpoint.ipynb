# Particle position vs. B (Particle histograms)

In this script, we will analyze the particle histograms along the whole experiment in order to resolve if the trajectories of particles are bistable or not. To do that, we will need to perform the analysis with a high statistic.

## 0. Functions involved in the process:

```python
def indexes_v_and_h(trj):
    
    """ This function takes the first frame of the experiment in order to filter which long traps
    are or vertical or horizontal."""
    
    frame = trj.query('frame==0')
    particles = frame.query('type==1')
    traps = frame.query('abs(mux)>10 or abs(muy)>10') # long traps
    ids = ids_long_traps(particles)
    mask = particles.index.get_level_values('id').isin(ids)
    atoms = particles.loc[mask]
   
    atoms = atoms.reset_index()
    traps = traps.reset_index()
    
    col = atoms-traps
    
    vertical = col.query('abs(mux) == 0')
    horizontal = col.query('abs(muy) == 0')

    v = vertical.index.get_level_values(0).values
    h = horizontal.index.get_level_values(0).values
    
    return v, h
```

Due to the lattice definition function particles id-labeled with $id = 1+nÂ·5$, $n = Z$ positives, are the particles located at long traps of the shakti geometry.

```python
def ids_long_traps(particles):
    
    """ In the following function we build an array of the id values of long traps in a shakti lattice """
    
    init = particles.index.get_level_values('id').min()
    fin = particles.index.get_level_values('id').max()
    ids = np.arange(init,fin,5)
    
    return ids
```

```python
def obtain_hist( trj,last_frame = 250000):
    
    """ I will do the histogram at a certain frame value """
    
    frame = trj.query('frame=='+str(last_frame))
    particles = frame.query('type==1')
    traps = frame.query('abs(mux)>10 or abs(muy)>10') # long traps
    ids = ids_long_traps(particles)
    
    mask = particles.index.get_level_values('id').isin(ids)
    atoms = particles.loc[mask]
    v, h = indexes_v_and_h(trj)
    
    #Once we only have the particles belonging to the long traps and
    #the long traps centers we can match each particle to the corresponding 
    #trap by looking for the nearesth neighbour trap/particle. 
    #It results that this is not necessary.
    #Traps and particles are ordered the same way.
    
    atoms = atoms.reset_index()
    traps = traps.reset_index()
    col = atoms-traps
    
    mask_v = col.index.get_level_values(0).isin(v)
    v = col.loc[mask_v]

    mask_h = col.index.get_level_values(0).isin(h)
    h = col.loc[mask_h]
    
    return traps, atoms, v, h
```

## End functions definition

```python
v_arr = []
h_arr =[]
for i in range(0,10):
    
    load_obj = mgc.trj_lazyread(os.path.join(data_directory,"ACI_ShaktiLattice_25mT-2500s_FreeBound_l10_exp"+str(i)+".lammpstrj"),
                            output = ["x","y","z","mux","muy","muz"]) 
    trj = load_obj.read_trj()
    print(i)
    # Perform analysis and append
    traps, atoms, v , h = obtain_hist(trj, 250000)
    v_arr.append(v.copy())
    h_arr.append(h.copy())
```
