# Simulation Shakti lattice

In this script, I will summarize the code needed to obtain the output for the shakti simulations. This is a previous step for having a script with parallel simulations.

```python
output = "/home/carolina/CairoLattice_ACI_PROJECT/Scripts/Output_Script-11.0"
```

```python
# <Define the geometric parameters of the lattice and build the spin lattice>

trap_sep = 10 # Short trap
radius = 5
distance = 3
interparticle_distance = 2*radius + distance # distance among a particle next to the other 
part_d = interparticle_distance*np.cos(np.pi/4) # distance among a particle in front of the other 
plaquette_cte = part_d*4+2*trap_sep
plaquette_cte/2

sp = sha.spins()
ureg = ice.ureg

sp.create_lattice(geometry = "shakti",size =[10,10],lattice_constant=plaquette_cte*ureg.um , border="periodic")
sp.order_spins(ordering.random_ordering)

# <Choose colloidal and trap parameters>

particle = ice.particle(radius = 5*ureg.um,
             susceptibility = 0.005,
             diffusion = 0.125*ureg.um**2/ureg.s,
             temperature = 300*ureg.K,
             density = 1000*ureg.kg/ureg.m**3)

trap_short = ice.trap(trap_sep = 10*ureg.um,
               height = 25*ureg.pN*ureg.nm,
               stiffness = 0.1*ureg.fN/ureg.nm)

trap_long = ice.trap(trap_sep =(20+13*np.sqrt(2))*ureg.um,
               height = 25*ureg.pN*ureg.nm,
               stiffness = 0.1*ureg.fN/ureg.nm)

# <Assign trap properties-position>

traps = []

for s in sp:
    
    if ( s.direction[0].magnitude >= plaquette_cte or s.direction[0].magnitude <= -plaquette_cte or 
        s.direction[1].magnitude >= plaquette_cte or s.direction[1].magnitude <= -plaquette_cte):
        
        traps.append(trap_long)
        
    else:
        
        traps.append(trap_short)

# <Generate Colloidal Ice>

col = ice.colloidal_ice(sp, particle, traps, height_spread = 0, susceptibility_spread = 0.1)
col.pad_region(5*ureg.um)

f, (ax1) = plt.subplots(1,1,figsize = (15,15))

col.display(ax1)
```

```python
# <Choose simulation parameters>

world = ice.world(
    field = 25*ureg.mT,
    temperature = 300*ureg.K,
    dipole_cutoff = 200*ureg.um)
#   boundaries = ["p", "p", "p"])

change_m_time = 2000*ureg.s
total_time = 2500*ureg.s
col.simulate(world,
             name = "shakti_10x10_25mT-TEST",
             include_timestamp = False,
             targetdir = output,
             framerate = 1*ureg.Hz,
             timestep = 10*ureg.ms,
             run_time = total_time,
             output = ["x","y","z","mux","muy","muz"])

col.sim.field.fieldz = "(time<2000000000)*v_Bmag*time/%f+(time>=2000000000)*v_Bmag"%(change_m_time.to(ureg.us).magnitude)

# <Run and load simulation>

col.run_simulation() 
col.load_simulation()

f, (ax1) = plt.subplots(1,1,figsize = (15,15))

col.set_state_from_frame(-1)
col.display(ax1)
```

```python
# <Upload vertices files topology and generate vertex array>

edges_shakti_values = pd.read_csv('edges_shakti_10x10.dat')
vertices = pd.read_csv('vertices_shakti_10x10.dat')

v_non_inferred = ice.vertices(
    positions = vertices[["x","y"]].values,
    edges = edges_shakti_values[["start","stop"]].values)
```

Let's generate a video of each frame in order to understant how the system evolves. 

```python
def count_vertices(vrt, column = "charge", column1 = "coordination"):
    vrt_count = vrt.groupby([column,column1]).count().iloc[:,0].to_frame("number")

    vrt_count = vrt_count.reindex(pd.MultiIndex.from_product(
        [vrt_count.index.get_level_values(column).unique(),
        vrt_count.index.get_level_values(column1).unique()]), fill_value = 0)
    
    total_vrt = vrt_count.groupby([column1]).sum()
    vrt_count["fraction"] = vrt_count.number / total_vrt.number
        
    return vrt_count
```

```python
def get_indices_separate_traps():

# <This loop will have the indexes of the colloids in the long and short traps>
# <The colloidal ice must be in the initial cnfiguration to filter the colloids by position>

    col.set_state_from_frame(0);

    variable_v = []
    variable_h = []

    fixed_v = []
    fixed_h = []
    i = 0

    for c in col:
        if (c.colloid[0].magnitude > 10 or c.colloid[0].magnitude < -10):
            variable_h.append(i)

        elif (c.colloid[1].magnitude > 10 or c.colloid[1].magnitude < -10):
            variable_v.append(i)

        elif (c.colloid[1].magnitude > 2.5 or c.colloid[1].magnitude < -2.5):
            fixed_v.append(i)

        elif (c.colloid[0].magnitude > 2.5 or c.colloid[0].magnitude < -2.5):
            fixed_h.append(i)
        i = i + 1
        
    return variable_h, variable_v, fixed_v, fixed_h
```

```python
def separate_traps(variable_h, variable_v, fixed_v, fixed_h, col):

    var_h = []
    var_v = []

    fix_h = []
    fix_v = []

    for i in range(0,len(col)):

        if i in variable_h:
            var_h.append(col[i].colloid.magnitude) 
        elif i in variable_v:
            var_v.append(col[i].colloid.magnitude)
        elif i in fixed_v:
            fix_v.append(col[i].colloid.magnitude)
        else:
            fix_h.append(col[i].colloid.magnitude)

    var_h = np.array(var_h)
    var_v = np.array(var_v)
    fix_h = np.array(fix_h)
    fix_v = np.array(fix_v)
    
    var = np.concatenate((var_h[:,0],var_v[:,1]))
    fix = np.concatenate((fix_h[:,0],fix_v[:,1]))
    
    return var , fix
```

```python
variable_h, variable_v, fixed_v, fixed_h = get_indices_separate_traps()
```

```python
var, fix = separate_traps(variable_h, variable_v, fixed_v, fixed_h, col)
```

```python
frames = col.trj.index.get_level_values("frame").unique()

v_df = []

for f in tqdm.tqdm(frames[::10]):
    col.set_state_from_frame(f)

    v_non_inferred.update_directions(col)
    v_non_inferred.classify_vertices(ice.ice_to_spins(col))
    vertices = v_non_inferred.vertices
    
    # <Leave only the window shown in the video>
    
    x_min = 200
    y_min = 200

    x_max = 600
    y_max = 600

    Ca = vertices.loc[(vertices['x']<x_max) & (vertices['x']>x_min)]
    Lattices_exp = Ca.loc[(Ca['y']<y_max) & (Ca['y']>y_min)]

    vrt_count = count_vertices(Lattices_exp)

    v_df.append(vrt_count.copy())


v_df = pd.concat(v_df,keys=frames[::10], names = ["frame"])
v_df = v_df.query('fraction != 0')
```

```python
def convert_frame_to_B(frame):
    rate = 25/2000
    if frame < 2000:
        field = frame*rate
    else:
        field = 25
    return str(round(field,1))
```

```python
frames = col.trj.index.get_level_values("frame").unique()

for f in tqdm.tqdm(frames[::10]):
    
    col.set_state_from_frame(f)
    var, fix = separate_traps(variable_h, variable_v, fixed_v, fixed_h, col)
    v_non_inferred.update_directions(col)
    v_non_inferred.classify_vertices(ice.ice_to_spins(col))

    fig, (ax1,ax2) = plt.subplots(2,2,figsize = (16,10), gridspec_kw={'height_ratios': [2, 0.5]})
    plt.subplots_adjust(wspace= 0.35, hspace = 0.6)
    
    field = convert_frame_to_B(f)

    col.display(ax1[0])
    ax1[0].set_title(r'$ \theta = 0° $' , size = 50)
    ax1[0].set_xlim(200, 600)
    ax1[0].set_ylim(200, 600)
    ax1[0].text(130, -20, s = field+' mT', size = 50)
    ax1[0].axis('off')
    v_non_inferred.display(ax1[0])    
    v_def = v_df.query('frame <= '+str(f)) 
    
    
    for typ_m, mtype in v_def.groupby(['charge']):
    
        ax1[1].plot(mtype.index.get_level_values('frame'), mtype["fraction"],'o-', label = typ_m, markersize = 5)

        ax1[1].tick_params(axis='y', labelsize=30)
        ax1[1].tick_params(axis='x', labelsize=30)  
        ax1[1].set_xlim(0,2500)
        ax1[1].set_ylim(0,1)
        ax1[1].legend(title = 'Charge',title_fontsize = 25, fontsize = 20, bbox_to_anchor=(1.3, 1.05))
        ax1[1].set_xlabel('time (s)', size = 40)
        ax1[1].set_ylabel('Charge population', size = 40)
        
    
    ax2[0].tick_params(axis='y', labelsize=30)
    ax2[0].tick_params(axis='x', labelsize=30) 
    ax2[1].tick_params(axis='y', labelsize=30)
    ax2[1].tick_params(axis='x', labelsize=30) 
    
    ax2[0].set_title('variable trap, 38.4 $ \mu m$' , size = 32)
    ax2[1].set_title(r'fixed trap, 10 $ \mu m$' , size = 32)
    ax2[0].hist(var, density = True, stacked = True, color = 'slateblue')
    ax2[1].hist(fix, density = True, stacked = True, color = 'slateblue')
    ax2[0].set_xlim(-20,20)
    #ax2[0].set_ylim(0,0.35)
    ax2[1].set_xlim(-5.5,5.5)
    #ax2[1].set_ylim(0,0.6)
    
    ax2[0].yaxis.set_major_formatter(FormatStrFormatter('%.1f'))
    ax2[1].yaxis.set_major_formatter(FormatStrFormatter('%.1f'))
    
    
    plt.savefig(str(f)+'_0', bbox_inches='tight', dpi = 100)
    plt.show()
```
