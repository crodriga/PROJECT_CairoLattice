Defining the elementary parameters to generate the lattice

```python
d = 13 # Interparticle distance
trap_sep_l = 10 # trap_sep parameter of the edge connecting 3 coordination and 4 coordination vertices
s_l4 = d*np.sqrt(2)/2
theta = 21*np.pi/180
```

```python
trap_sep_a, s_l3, a, l, plaquette_cte = lattices.lattice_parameters(theta, d, trap_sep_l, s_l4)
```

```python
sp = lattices.spins()
ureg = ice.ureg
Sx = 5
Sy = 5
sp.create_lattice(geometry = "from_cairo_to_shakti", size =[Sx,Sy], theta = theta,d = d ,trap_sep_l = trap_sep_l ,s_l4 = s_l4, plaquette_cte = plaquette_cte*ureg.um, border="periodic")
sp.order_spins(ordering.random_ordering)
```

```python
particle = ice.particle(radius = 5*ureg.um,
             susceptibility = 0.005,
             diffusion = 0.125*ureg.um**2/ureg.s,
             temperature = 300*ureg.K,
             density = 1000*ureg.kg/ureg.m**3)

trap_long = ice.trap(trap_sep = trap_sep_l*ureg.um,
               height = 30*ureg.pN*ureg.nm,
               stiffness = 0.1*ureg.fN/ureg.nm)

trap_short = ice.trap(trap_sep = trap_sep_a*ureg.um,
               height = 30*ureg.pN*ureg.nm,
               stiffness = 0.1*ureg.fN/ureg.nm)

traps = []
```

```python
for s in sp:
    
    if ( s.direction[0].magnitude == 2*trap_sep_l or s.direction[0].magnitude == -2*trap_sep_l or 
        s.direction[1].magnitude == 2*trap_sep_l or s.direction[1].magnitude == -2*trap_sep_l ):
        
        traps.append(trap_short)
        
    else:
        
        traps.append(trap_long)
```

```python
col = ice.colloidal_ice(sp, particle, traps, height_spread = 0, susceptibility_spread = 0.1)

col.pad_region(20*ureg.um) 
```

```python
def new_display(col, ax1=False, units = None, scale = 1):
    """ Draws a figure with the trap and the colloid inside it"""
    if not ax1:
        fig, ax1 = plt.subplots(1,1)
        
    for c in col:
        patches = new_create_patch(c,units = units, scale=scale)

        ax1.add_patch(patches[0])
        ax1.add_patch(patches[1])
        ax1.add_patch(patches[2])
        ax1.add_patch(patches[3])
        ax1.add_patch(patches[4])
    
def new_create_patch(col, units = None, scale = 1):
    """ Draws a figure with the trap and the colloid inside it"""

    if not units:
        units = col.center.units

    X=col.center[0].to(units).magnitude
    Y=col.center[1].to(units).magnitude

    # D is the vector that goes fom the center to each of the traps
    DX=col.direction[0]/2*col.trap.trap_sep.to(units).magnitude
    DY=col.direction[1]/2*col.trap.trap_sep.to(units).magnitude
    # P is the vector that goes from the center to the colloid
    PX=col.colloid[0].to(units).magnitude
    PY=col.colloid[1].to(units).magnitude
    # B is the euclidean distance between minimums
    B = DX**2+DY**2
    # A is the angle formed with the x-axis
    A = np.arccos(2*DX/B)*180/np.pi
    #print(A)
    W = (col.particle.radius.to(units).magnitude)

    return [Circle((X-DX,Y-DY), radius = W*scale, ec='g', fc='green'),
            Circle((X+DX,Y+DY), radius = W*scale, ec='y', fc='green'),
            Circle((DX,DY), radius = W*scale, ec='y', fc='red'),
            Rectangle((X-DX,Y-DY), width = B, height =4*W*scale , angle = 0, fc = 'black'),
            Circle((X+PX,Y+PY), radius = W, ec='k', fc = 'none')]
```

```python
frames = col.trj.index.get_level_values("frame").unique()

v_df = []

for f in tqdm.tqdm(frames[::50]):
    col.set_state_from_frame(f)

    v_non_inferred.update_directions(col)
    v_non_inferred.classify_vertices(ice.ice_to_spins(col))
    vertices = v_non_inferred.vertices

    v_df.append(vertices.copy())


v_df = pd.concat(v_df,keys=frames[::50], names = ["frame"])
```

```python
First_frame = v_df.loc[0]
```

```python
x_min = plaquette_cte/2
y_min = plaquette_cte/2

width = plaquette_cte*7
hight = plaquette_cte*7

x_max = x_min+width
y_max = y_min+hight

print(x_min,x_max,y_min,y_max)

fig =plt.figure(figsize = (15,15))
ax = fig.add_subplot(111)

plt.plot(First_frame.x, First_frame.y , 'o')
ax.add_patch(Rectangle((x_min, y_min),width, hight,fc ='none', ec ='g',lw = 5))

Ca = v_df.loc[(v_df['x']<x_max) & (v_df['x']>x_min)]
Lattices_exp = Ca.loc[(Ca['y']<y_max) & (Ca['y']>y_min)]
```
